```python
start = 50
dictionnary = []
```

```python
with open("input.txt", "r") as puzzle_input:
    for line in puzzle_input:
        dictionnary.append([line[0], int(line[1:].strip())])
```

```python
dictionnary
```

```python
for line in dictionnary:
    print(line[0])
```

```python
for line in dictionnary:
    if line[0] == 'L':
        line[1] = -line[1]

print(dictionnary)
```

```python
counter = start
zero_counter = 0

for line in dictionnary:
    counter = (counter + line[1]) % 100
    print(counter)
    if counter == 0:
        zero_counter += 1
```
